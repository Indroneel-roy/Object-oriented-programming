###Why use self in class

```python
class Dog:
    def __init__(self, name, breed):
        self.name = name  # self.name refers to the 'name' attribute of the specific dog object
        self.breed = breed

    def bark(self):
        print(f"{self.name} says Woof!")

# Create two dog objects
dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Lucy", "Labrador")

# Call the bark method for each dog
dog1.bark()  # Output: Buddy says Woof! (self refers to dog1)
dog2.bark()  # Output: Lucy says Woof! (self refers to dog2)
```

In this example:

When dog1.bark() is called, self in the bark method refers to dog1. So, it prints "Buddy says Woof!".

When dog2.bark() is called, self refers to dog2. So, it prints "Lucy says Woof!".

Without self, the bark method wouldn't know which dog's name to print.

In summary, self is crucial for making object-oriented programming work in Python by ensuring that methods operate on the correct object's data.

Fraction Class

```python
class Fraction:
    def __init__(self, n, d):
        self.neu = n
        self.deno = d
    def __str__(self):
        return "{}/{}".format(self.neu, self.deno)
    def __add__(self, other):
        tem_neu = self.neu * other.deno + self.deno * other.neu
        tem_deno = self.deno * other.deno
        return "{}/{}".format(tem_neu, tem_deno)
    def __sub__(self, other):
        tem_neu = self.neu * other.deno - self.deno * other.neu
        tem_deno = self.deno * other.deno
        return "{}/{}".format(tem_neu, tem_deno)
    def __mul__(self, other):
        tem_neu = self.neu * other.neu
        tem_deno = self.deno * other.deno
        return "{}/{}".format(tem_neu, tem_deno)
    def __truediv__(self, other):
        tem_neu = self.neu * other.deno
        tem_deno = self.deno * other.neu
        return "{}/{}".format(tem_neu, tem_deno)
```

```python
p = Fraction(5, 4)
print(p)
```

```python
p = Fraction(5, 4)
q = Fraction(5, 3)

print(p + q )
print(p - q)
print(p/q)
print(p * q)
```

###Polymorphism

```python
class Phone:
  def __init__(self, price, brand, camera):
    self.price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
    print("Buying a phone")
class SmartPhone(Phone):
  def buy(self):
    print("Buying a smartphone")

```

```python
a = SmartPhone(20000, "Samsung", 13)
a.buy() # there is a same name method in both class but when you call the buy it will conduct the inside buy method this is call method overwriting
```

```python
### Use of super()
class Phone:
  def __init__(self, price, brand, camera):
    self.price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
    print("Buying a phone")
class SmartPhone(Phone):
  def buy(self):
    print("Buying a smartphone")
    super().buy()
```

```python
a = SmartPhone(20000, "Samsung", 13)
a.buy()
```
